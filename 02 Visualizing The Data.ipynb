**Deals With The Warning** , It is because we fork the original data itself, so rather than fork the data, we can create a copy of a data sets from original datasets instead.

```python
df_albany = df.copy()[ df["region"] == "Albany"] # This is the new copy data
df_albany
```

```python
df_albany.set_index("Date")
df_albany.sort_index(inplace=True)
df_albany["PriceMean25"] = df_albany["AveragePrice"].rolling(25).mean()
```

```python
import pandas as pd

df = pd.read_csv("datasets/avocado.csv")

# Convert The Date
df = df.copy()[ df["type"]=="organic"]
df["Date"] = pd.to_datetime(df["Date"])

graph_df = pd.DataFrame()


for region in df["region"].unique():
    print(region)
    df_region = df.copy()[ df["region"] == region ]
    df_region.set_index("Date", inplace=True)
    df_region.sort_index(inplace=True)
    df_region[f"{region}PriceMean25"] = df_region["AveragePrice"].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = df_region[[f"{region}PriceMean25"]]
    else:
        graph_df = graph_df.join(df_region[f"{region}PriceMean25"])
    
```

```python
graph_df.plot(figsize=(10,5), legend=False)
```
